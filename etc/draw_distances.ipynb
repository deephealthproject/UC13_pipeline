```python
import matplotlib.pyplot as plt
%matplotlib notebook
import pickle
import numpy as np
```

```python
with open('dist_normal_pacient02_model2big.pkl', 'rb') as f1:
    dist_normal = pickle.load(f1)
with open('dist_seizure_pacient02_model2big.pkl', 'rb') as f2:
    dist_seizure = pickle.load(f2)
```

```python
print(len(dist_normal))
print(len(dist_seizure))
print("Normal samples: Max->", max(dist_normal),
      "Min->", min(dist_normal), 
      "Avg->", sum(dist_normal)/len(dist_normal))
print("Seizure samples: Max->", max(dist_seizure),
      "Min->", min(dist_seizure), 
      "Avg->", sum(dist_seizure)/len(dist_seizure))
```

```python
plt.figure(figsize=(9,5))
plt.title("L2 reconstruction distance")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.plot(dist_normal,label="Normal periods")
plt.plot(dist_seizure,label="Seizure periods")
plt.legend()
plt.savefig("reconstruction_dist_pacient02.jpg")
plt.show()
```

```python
plt.figure(figsize=(9,5))
n, bins, patches = plt.hist( np.log(1 + np.array(dist_normal)), 100, density=False, facecolor='b', alpha=0.30 )
n, bins, patches = plt.hist( np.log(1 + np.array(dist_seizure)), 100, density=False, facecolor='r', alpha=0.30 )


plt.xlabel( 'Values' )
plt.ylabel( 'Probability' )
plt.title( 'Generic Example of Histogram' )
#pyplot.text( 60, 0.025, r'$\mu=%d,\ \sigma=%d$' % (mu, sigma) )
#plt.axis( [ 0, 160, 0, 0.1 ] )
plt.grid(True)
plt.show()
```
